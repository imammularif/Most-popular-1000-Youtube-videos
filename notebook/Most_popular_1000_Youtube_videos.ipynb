```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from matplotlib.ticker import FuncFormatter
```

```python
file_name = "Most popular 1000 Youtube videos.csv"
```

```python
# Baca file
df = pd.read_csv(file_name)
```

```python
# Bersihkan angka dari koma dan ubah ke float
for col in ['Video views', 'Likes', 'Dislikes']:
    df[col] = df[col].replace({',': ''}, regex=True).astype(float)
```

TOTAL VIEW, LIKES, DISLIKES, & JUMLAH RANK

```python
# Hitung total
total_views = df['Video views'].sum()
total_likes = df['Likes'].sum()
total_dislikes = df['Dislikes'].sum()
total_rank = len(df)

# Print hasil
print(f"Total Views: {total_views:,.0f}")
print(f"Total Likes: {total_likes:,.0f}")
print(f"Total Dislikes: {total_dislikes:,.0f}")
print(f"Total Rank (Data): {total_rank:,}")
```

MENAMPILKAN RANK, VIDIO, LIKES, DISLIKES, VIDIO VIEWS, & PUBLISHED

```python
cols = ['rank', 'Video', 'Likes', 'Dislikes', 'Video views', 'published']
df_clean = df[cols]

df_clean.head(10)
```

Urutkan dari yang TERBANYAK views (top 10)

```python
top10 = df_clean.sort_values(by='Video views', ascending=False).head(10)
print("10 Video dengan Views Terbanyak")
display(top10)
```

Urutkan dari yang TERENDAH views (bottom 10)

```python
bottom10 = df_clean.sort_values(by='Video views', ascending=True).head(10)
print("10 Video dengan Views Terendah")
display(bottom10)
```

DONUTS CHAR

```python
# --- Hitung jumlah tiap kategori
category_counts = df['Category'].value_counts()
```

```python
# --- Hitung persentase
percentages = (category_counts / category_counts.sum()) * 100
```

```python
# --- (Warna opsional)
colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#bab0ac']
```

```python
# --- Ambil top 5 + ‚ÄúLainnya‚Äù
top5 = category_counts.head(5)
others = pd.Series({'Lainnya': category_counts.iloc[5:].sum()})
final_category = pd.concat([top5, others])

# --- Deskripsi kategori
category_desc = {
    'Music': 'Musik & konser',
    'Entertainment': 'Hiburan & film',
    'People & Blogs': 'Vlog & cerita pribadi',
    'Comedy': 'Konten lucu/parodi',
    'Gaming': 'Game & streaming',
    'Lainnya': 'Kategori kecil lainnya'
}

# --- DataFrame
cat_df = pd.DataFrame({
    'Category': final_category.index,
    'Count': final_category.values
})
cat_df['Deskripsi'] = cat_df['Category'].map(category_desc)

# --- Warna lembut dashboard
colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#bab0ac']

# --- Chart
fig = go.Figure()

fig.add_trace(go.Pie(
    labels=cat_df['Category'],
    values=cat_df['Count'],
    hole=0.55,
    marker=dict(colors=colors, line=dict(color='white', width=1)),
    textinfo='percent+label',
    textposition='outside',
    showlegend=False
))

# --- bullet text
annotations = []
y_pos = 0.85
for i, row in enumerate(cat_df.itertuples()):
    annotations.append(dict(
        x=1.20,  # posisi kanan chart
        y=y_pos - i*0.15,  # jarak antar baris
        xanchor='left',
        text=f"<b><span style='color:{colors[i]}'>‚¨§</span> {row.Category}</b><br><span style='font-size:11px;color:#555;'>{row.Deskripsi}</span>",
        showarrow=False,
        align='left'
    ))

# --- Layout
fig.update_layout(
    width=750,
    height=600,
    margin=dict(t=80, b=40, l=40, r=250),
    annotations=annotations,
    paper_bgcolor='white',
    plot_bgcolor='white',
    title_text='Category menurut Record Count',
    title_font=dict(size=18, family='Arial Black'),
)


fig.show()
```

BAR CHART

```python
# --- Get Column ---
df = df[['rank', 'Video views']].dropna()
df['Video views'] = pd.to_numeric(df['Video views'], errors='coerce')
```

```python
# --- Ambil 5 teratas berdasarkan views ---
top5 = df.nlargest(5, 'Video views')

# --- Format angka jadi jutaan (misal 6.547.981.039 ‚Üí 6.5 B) ---
def format_milions(x, pos):
    if x >= 1e9:
        return f'{x/1e9:.1f} B'
    elif x >= 1e6:
        return f'{x/1e6:.1f} M'
    return int(x)

# --- Warna biru ---
color = '#4285F4'

# --- Bikin figure ---
fig, ax = plt.subplots(figsize=(10, 4.5))
bars = ax.barh(top5['rank'].astype(str), top5['Video views'], color=color)

# --- Tambah label di batangnya ---
for bar in bars:
    value = bar.get_width()
    ax.text(value - (value * 0.25), bar.get_y() + bar.get_height()/2,
            f"{value:,.0f}", ha='right', va='center', color='white', fontsize=10, fontweight='bold')

# --- Styling chart ---
ax.set_title('Video views menurut rank', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Video views', fontsize=11)
ax.set_ylabel('rank', fontsize=11)
ax.xaxis.set_major_formatter(FuncFormatter(format_milions))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# --- Tambah border oranye ---
for spine in ax.spines.values():
    spine.set_edgecolor('#f28e2b')
    spine.set_linewidth(2)

plt.tight_layout()
plt.show()
```
